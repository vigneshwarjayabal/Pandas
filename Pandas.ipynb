#**Filter Operation in pandas**

image.png

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
df=pd.read_csv('/content/Car_sales.csv')
df
```

```python
df[df["Manufacturer"]=="Toyota"] # this filters data where the value in the Manufacturer column is equal to "Toyota".
```

```python
df[df["Vehicle_type"]=="Passenger"]
```

```python
# used AND Operation in filtering and you can also use & insdeat of  AND.
df[(df["Manufacturer"]=="Toyota") & (df["Vehicle_type"]=="Passenger")]
```

```python
 # used OR operation in filtering and you can used | instead of OR.
df[(df["Manufacturer"]=="Toyota") | (df["Vehicle_type"]=="Passenger")]
```

```python
df[df["Horsepower"]>=200]
```

```python
 # between function() will return all data within the given range .
df[df["Horsepower"].between(200,220)]
```

```python
# The value_counts() method in pandas is used to count the occurrences of unique values in a Series (single column).
df["Vehicle_type"].value_counts()
```

```python
df[df["Horsepower"].between(200,220)]["Vehicle_type"].value_counts()
```

```python
# filter data using isin() function
#The isin() method checks if the Dataframe contains the specified value(s).
df[df["Manufacturer"].isin(["Toyota","Honda"])]
```

```python
df[df["Manufacturer"].isin(["Toyota","Honda"])].loc[:,["Engine_size","Wheelbase"]]
```

```python
df[df["Manufacturer"].isin(["Toyota","Honda"])].value_counts("Manufacturer")
```

```python
df[df["Manufacturer"].isin(["Toyota","Honda"])].value_counts()
```

#**Sort() Function in pandas**

image.png

```python
df.sort_values(by=["Manufacturer"],ascending=False) # ascending=false:will sort in descending order, if it is true it will sort in ascending.
```

```python
df.sort_values(by=["Manufacturer"],ascending=True)
```

```python
df.sort_values(by=["Manufacturer","Vehicle_type","Horsepower","Price_in_thousands"],ascending=[False,True,False,True])
```

```python
df.sort_values(by=["Manufacturer","Vehicle_type","Horsepower","Price_in_thousands"],ascending=[False,True,False,True]).loc[:,["Manufacturer","Vehicle_type","Horsepower","Price_in_thousands"]]
```
