#**Set_index() function in pandas**

```python
df_car.set_index("Length") # setting a row index with values in Length label
```

set_index() fuction well temparvaryly set a row index value with column value we specify.

if you add inplace=True then it will permentaly set  a row index value with column value we specify.

```python
df_car.iloc[-4:-1] # negative index start from the last to first
```
